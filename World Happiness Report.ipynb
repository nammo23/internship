```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

from google.colab import files 
uploaded = files.upload()

data = pd.read_csv('World_Happiness_2015_2017_.csv')
data.head()
data.shape
data.dtypes
data['Year'] = data['Year'].astype('float')
data.dtypes
data.head()
data['Year'] = data['Year'].astype('int')
data.head()
data.isnull().sum()
g = sns.pairplot(data)
g.fig.suptitle('FacetGrid plot', fontsize = 20)
g.fig.subplots_adjust(top= 0.9);
     
eco_happiness = ['Happiness Score','Economy (GDP per Capita)']
eco_corr = data[eco_happiness]
eco_corr.corr()
sns.regplot(data = eco_corr, x = 'Happiness Score', y = 'Economy (GDP per Capita)').set_title("Emini's Correlation graph for Happiness score vs Economy")

fami_happiness = ['Happiness Score','Family']
fami_corr = data[fami_happiness]
fami_corr.corr()
sns.regplot(data = fami_corr, x = 'Happiness Score', y = 'Family').set_title("Emini's Correlation graph for Happiness score vs Family")

hea_happiness = ['Health (Life Expectancy)','Happiness Score']
hea_corr = data[hea_happiness]
hea_corr.corr() 
sns.regplot(data = hea_corr, x = 'Happiness Score', y = 'Health (Life Expectancy)').set_title("Emini's Correlation graph for Happiness score vs Health")

free_happiness = ['Freedom', 'Happiness Score']
free_corr = data[free_happiness]
free_corr.corr()
sns.regplot(data = free_corr, x = 'Happiness Score', y = 'Freedom').set_title("Emini's Correlation graph for Happiness score vs Freedom")

trust_happiness = ['Trust (Government Corruption)','Happiness Score']
trust_corr = data[trust_happiness]
trust_corr.corr()
sns.regplot(data = trust_corr, x = 'Happiness Score', y = 'Trust (Government Corruption)').set_title("Emini's Correlation graph for Happiness score vs Gov. Corruption")

gene_happiness = ['Generosity','Happiness Score']
gene_corr = data[gene_happiness]
gene_corr.corr()
sns.regplot(data = gene_corr, x = 'Happiness Score', y = 'Generosity').set_title("Emini's Correlation graph for Happiness score vs Generosity")

dys_happiness = ['Dystopia Residual','Happiness Score']
dys_corr = data[dys_happiness]
dys_corr.corr()
sns.regplot(data = dys_corr, x = 'Happiness Score', y = 'Dystopia Residual').set_title("Emini's Correlation graph for Happiness score vs Dystopia Residual")
     
```
